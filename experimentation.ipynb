# Adaptative Hibrid Graph Convolutional Network (AH-GCN).

Este notebook tiene como objetivo explorar y experimentar con una arquitectura propuesta (una modificación de la GCN) para ver si puede aprender bien tanto en grafos homofílicos (donde la GCN suele funcionar bien) como en heterofílicos (donde no suele funcionar bien). La descripción de la arquitectura esta más detallada en la memoria, pero esta es la idea:

$$X^{(l+1)} = \sigma \left( \alpha^{(l)} S_{LP} X^{(l)} W^{(l)} + (1 - \alpha^{(l)}) S_{HP} X^{(l)} W^{(l)} \right)$$

Como se puede apreciar, es muy similar a la GCN. La modificación propuesta es, en cada capa, combinar dos GSO distintos, uno *low pass* y otro *high pass*, con un regulador $\alpha$, que idealmente se aprendera por la red para que esta le pueda dar mas importancia a las frecuencias bajas (homofilia) o a las altas (heterofilia). Además, la idea también es que la red, en caso de que sea necesario, pueda saturar el $\alpha$ a uno y así recupere perfectamente una GCN estandar.
